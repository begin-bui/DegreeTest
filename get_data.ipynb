## ST数据

```python
st_data = pd.read_csv('st_data_2021-2022.csv', index_col=False)
st_data['date'] = st_data['date'].apply(lambda x: datetime.strptime(str(x), '%Y-%m-%d'))
st_data.set_index(['asset','date'])
## 其中st为是否st股票的标识，1表示该股票当天为st，0表示该股票当天不为st
st_data.head()
```

## 行业数据

```python
# 行业数据提供中信一级行业
indus_data = pd.read_csv('indus_data_2021-2022.csv', index_col=False)
indus_data['date'] = indus_data['date'].apply(lambda x: datetime.strptime(str(x), '%Y-%m-%d'))
indus_data.set_index(['asset','date'])
# 其中hjcode为股票代码，first_industry_code为股票当日所属中信一级行业代码，first_industry_name为股票当日所属中信一级行业名称，trans_date为交易日期
indus_data.head()
```

## 行情数据

```python
kline_data = pd.read_csv('kline_data_2021-2022.csv', index_col=False)
kline_data['date'] = kline_data['date'].apply(lambda x: datetime.strptime(str(x), '%Y-%m-%d'))
kline_data.set_index(['asset','date'])
"""
注：价格数据为了保存精度统一乘10000，即实际价格为表中价格/10000
hjcode:股票代码
market:市场
open_price:开盘价
high_price:最高价
low_price:最低价
close_price:收盘价
turnover:成交金额
volume:成交量
trans_date:交易日
"""
kline_data.head()
```
